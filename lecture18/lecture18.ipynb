# Data formats on the web

## JSON

````python
import json

json_text = """
{
    "note": {
        "to": "Tove",
        "from": "Jani",
        "heading": "Reminder",
        "body": "Don't forget me this weekend!"
    }
}
"""

note_dict = json.loads(json_text)
print('* Converted to dict')
print(note_dict)
note_json = json.dumps(note_dict, indent=4)
print('* JSON format')
print(note_json)
````

## XML

Here I use the following two packages, which gives similiar interfaces to the `json` package.
- `xmltodict`
- `dicttoxml`

For the full-featured xml package, use the `xml` package.

````python
import xmltodict
import dicttoxml
# for prettier priting
import pprint

print('* Simple XML format')
print(dicttoxml.dicttoxml(note_dict, root=False, attr_type=False))
print('* A little bit complex XML format')
print(dicttoxml.dicttoxml(note_dict))

simple_xml_text = dicttoxml.dicttoxml(note_dict, root=False, attr_type=False)
print('* OrderedDict')
print(xmltodict.parse(simple_xml_text))
print('* Dict')
print(dict(xmltodict.parse(simple_xml_text)['note']))
````

## Kakao Map APIs

Documentation: https://developers.kakao.com/docs/restapi/local

Example of querying locations with a keyword: 

```
curl -v -X GET "https://dapi.kakao.com/v2/local/search/keyword.json?y=37.514322572335935&x=127.06283102249932&radius=20000" \
--data-urlencode "query=카카오프렌즈" \
-H "Authorization: KakaoAK kkkkkkkkkkkkkkkkkkkkkkkkkkkkkkkk"
```

````python
import requests
from bs4 import BeautifulSoup

url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
params = {
    'query': '디지스트'
}
headers = {
    'Authorization': f'KakaoAK {config.KAKAO_REST_API_KEY}'
}
r = requests.get(url, params=params, headers=headers)
````

````python
import json
response = json.loads(r.text)
response
````

````python
for place in response['documents']:
    print(place['place_name'], ':', place['road_address_name'])
````

# Presentation Order

## Import packages

```python
import numpy as np
import pandas as pd
import datetime

np.random.seed(377)  # well, using 377 seems reasonable...
```

## Provide configurations 

```python
n_teams = 22
dates = [datetime.date(2017, 12, 7),
         datetime.date(2017, 12, 11),
         datetime.date(2017, 12, 14),
         datetime.date(2017, 12, 15)]
allocatable = np.array([0, 7, 7, 7])

# pre-allocated teams: (team, date)
pre_allocation = [
    (2, datetime.date(2017, 12, 7)),
    (4, datetime.date(2017, 12, 7)),
    (11, datetime.date(2017, 12, 7)),
    (20, datetime.date(2017, 12, 7)),
]
```

## Initialize variables & check the feasibility

```python
teams_allocated = [team for team, date in pre_allocation]
teams_remaining = list(set(range(1, n_teams + 1)) - set(teams_allocated))

if len(teams_remaining) > sum(allocatable):
    print("THIS CANNOT BE ALLOCATABLE")
```

## Determine presentation orders

```python
# handle pre-determined teams
# set the order
presentations = np.random.permutation(pre_allocation)
# convert it to a list for easier manipulation later
presentations = [(team, date) for team, date in presentations]
```

```python
# now, handle remaining teams
teams_remaining = np.random.permutation(teams_remaining)

for i, date in enumerate(dates):
    # choose teams for a given date
    new_allocations = teams_remaining[:allocatable[i]]
    presentations += [(team, date) for team in new_allocations]
    teams_remaining = teams_remaining[allocatable[i]:]
```

## Print by team

```python
for team, date in sorted(presentations, key=lambda schedule: schedule[0]):
    print(f"{team:2}: {date.strftime('%m/%d')}")
```

## Print by date

```python
prev_date = None
for team, date in sorted(presentations, key=lambda schedule: schedule[1]):
    if prev_date != date:
        print('\n', date.strftime('%m/%d'), sep='', end=': ')
    prev_date = date
    print(f'{team:3}', end=' ')
```

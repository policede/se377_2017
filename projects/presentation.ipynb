# Presentation Order

## Package import

```python
import numpy as np
import pandas as pd
import datetime

np.random.seed(377)  # well, using 377 seems reasonable...
```

## Configurations 

```python
n_teams = 22
teams_per_day = 7

dates = [datetime.date(2017, 12, 7),
         datetime.date(2017, 12, 11),
         datetime.date(2017, 12, 14),
         datetime.date(2017, 12, 15)]
allocatable = np.array([False, True, True, True])

# pre-allocated teams: (team, date)
pre_allocation = [(2, datetime.date(2017, 12, 7)),
                  (4, datetime.date(2017, 12, 7)),
                  (11, datetime.date(2017, 12, 7)),
                  (20, datetime.date(2017, 12, 7)),
                 ]
```

## Initialization of variables, inferred from configuration

```python
allocated_dates = np.ndarray(n_teams + 1, dtype=object)  # to start index from 1
for team, date in pre_allocation:
    allocated_dates[team] = date
```

## Determine presentation orders

```python
for i, date in enumerate(dates):
    if not allocatable[i]:
        continue
        
    remaining_teams = np.where(allocated_dates == None)[0]  # use [0] since this is 1-d
    remaining_teams = remaining_teams[1:]  # use [1:] to remove team 0

    # quit for-loop if all teams are allocated
    if not len(remaining_teams):
        break
        
    if len(remaining_teams) < teams_per_day:
        new_allocation = remaining_teams
    else:
        new_allocation = np.random.choice(remaining_teams, teams_per_day, replace=False)
    allocated_dates[new_allocation] = date
```

## Print by team

```python
# use [1:] and start to make team number 1
for i, date in enumerate(allocated_dates[1:], start=1):
    print(i, date.strftime('%m/%d'))
```

## Print by date

```python
for date in dates:
    print(date.strftime('%m/%d'), ': ', sep='', end='')
    print(np.where(allocated_dates == date)[0])
```

## Additional Note

Lessons learned: Team number should start from 0 for cleaner and easier coding!

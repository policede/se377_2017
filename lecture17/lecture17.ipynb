```python
import requests
from bs4 import BeautifulSoup
```

## Basic web scraping

```python
# 망고플레이트: 대구 맛집 베스트 25곳
url = 'https://www.mangoplate.com/top_lists/1268_daegu'
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html5lib')
```

```python
soup
```

### Check how web page is structured

```python
soup.find_all('h3')
```

```python
soup.find_all('p', {'class': 'etc'})
```

### Retrieve information

```python
names = soup.find_all('h3')
addresses = soup.find_all('p', {'class': 'etc'})[:len(names)]

restaurants = {}
for name, address in zip(names, addresses):
    name = name.text.strip()
    address = address.text
    restaurants[name] = address
restaurants
```
